```python
import requests
from bs4 import BeautifulSoup
import json
from sentence_transformers import SentenceTransformer

#Collection page url
url = "YOUR WEBSITE URL"
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")
single_prd = soup.findAll(class_="product-card__info")
for prd in single_prd:
    title = prd.find("a").text
    prd_url = prd.find("a")["href"]
    
    html = requests.get(prd_url).text
    soup = BeautifulSoup(html, "html.parser")
    description = soup.find(class_="product-details__description").text


    data = [{
        "title": title,
        "description": description,
        "source_url": url
    }]

    embed_model = SentenceTransformer("all-MiniLM-L6-v2")

    for ch in data:
        
        vec = embed_model.encode(ch["description"]).tolist()
        payload = {
            
            "text": ch["description"],
            "embedding": json.dumps(vec),
            "source_url": ch["source_url"]
        }
    
    # Send to Laravel API
    api_url = "http://127.0.0.1:8000/api/docs"
    response = requests.post(api_url, json=payload)
    print(response.json())




```
