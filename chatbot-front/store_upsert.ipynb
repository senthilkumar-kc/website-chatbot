```python
import requests
import json
from sentence_transformers import SentenceTransformer
API = "http://127.0.0.1:8000/api/docs"
embed_model = SentenceTransformer("all-MiniLM-L6-v2")

chunks = [{'chunk_id':101, 'text':'Shape Your Future. Find Your Jobs', 'url':'https://yourwebsite.com/'}]

for ch in chunks:
    
    vec = embed_model.encode(ch["text"]).tolist()
    payload = {
        "chunk_id": ch["chunk_id"],
        "text": ch["text"],
        "embedding": json.dumps(vec),
        "source_url": ch["url"]
    }
    print(payload)
    res = requests.post(API, json=payload)
print(res)
```
